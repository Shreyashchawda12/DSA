```python
#1
class Node:
    def __init__(self, data=None):
        self.data = data
        self.next = None


def create_new_linked_list(head1, head2):
    if head1 is None and head2 is None:
        return None

    new_head = None
    new_tail = None

    while head1 is not None and head2 is not None:
        if head1.data >= head2.data:
            new_node = Node(head1.data)
            head1 = head1.next
        else:
            new_node = Node(head2.data)
            head2 = head2.next

        if new_head is None:
            new_head = new_node
            new_tail = new_node
        else:
            new_tail.next = new_node
            new_tail = new_node

    # Add remaining nodes from the first linked list
    while head1 is not None:
        new_node = Node(head1.data)
        new_tail.next = new_node
        new_tail = new_node
        head1 = head1.next

    # Add remaining nodes from the second linked list
    while head2 is not None:
        new_node = Node(head2.data)
        new_tail.next = new_node
        new_tail = new_node
        head2 = head2.next

    return new_head


# Example usage:
# Linked list 1: 1 -> 3 -> 5 -> 7 -> None
head1 = Node(1)
head1.next = Node(3)
head1.next.next = Node(5)
head1.next.next.next = Node(7)

# Linked list 2: 2 -> 4 -> 6 -> 8 -> None
head2 = Node(2)
head2.next = Node(4)
head2.next.next = Node(6)
head2.next.next.next = Node(8)

new_head = create_new_linked_list(head1, head2)

# Output: 2 -> 4 -> 6 -> 7 -> None
while new_head is not None:
    print(new_head.data, end=" -> ")
    new_head = new_head.next
print("None")
```

```python
#2
class Node:
    def __init__(self, data=None):
        self.data = data
        self.next = None


def remove_duplicates(head):
    if head is None:
        return None

    current = head

    while current.next is not None:
        if current.data == current.next.data:
            duplicate = current.next
            current.next = duplicate.next
            duplicate.next = None
        else:
            current = current.next

    return head


# Example usage:
# Linked list: 11 -> 11 -> 11 -> 21 -> 43 -> 43 -> 60 -> None
head = Node(11)
head.next = Node(11)
head.next.next = Node(11)
head.next.next.next = Node(21)
head.next.next.next.next = Node(43)
head.next.next.next.next.next = Node(43)
head.next.next.next.next.next.next = Node(60)

head = remove_duplicates(head)

# Output: 11 -> 21 -> 43 -> 60 -> None
while head is not None:
    print(head.data, end=" -> ")
    head = head.next
print("None")
```

```python
#4
class Node:
    def __init__(self, data=None):
        self.data = data
        self.next = None


def reverse_alternate_k_nodes(head, k, reverse):
    if head is None:
        return None

    count = 0
    current = head
    prev = None

    # Reverse k nodes if reverse is True, else skip k nodes
    while current is not None and count < k:
        if reverse:
            next_node = current.next
            current.next = prev
            prev = current
            current = next_node
        else:
            prev = current
            current = current.next

        count += 1

    # Recursive call to reverse the next alternate k nodes
    if reverse:
        head.next = reverse_alternate_k_nodes(current, k, not reverse)
        return prev
    else:
        prev.next = reverse_alternate_k_nodes(current, k, not reverse)
        return head


# Example usage:
# Linked list: 1 -> 2 -> 3 -> 4 -> 5 -> 6 -> 7 -> 8 -> None
head = Node(1)
head.next = Node(2)
head.next.next = Node(3)
head.next.next.next = Node(4)
head.next.next.next.next = Node(5)
head.next.next.next.next.next = Node(6)
head.next.next.next.next.next.next = Node(7)
head.next.next.next.next.next.next.next = Node(8)

k = 3
head = reverse_alternate_k_nodes(head, k, True)

# Output: 3 -> 2 -> 1 -> 4 -> 5 -> 6 -> 8 -> 7 -> None
while head is not None:
    print(head.data, end=" -> ")
    head = head.next
print("None")
```

```python
#5
class Node:
    def __init__(self, data=None):
        self.data = data
        self.next = None

def delete_last_occurrence(head, key):
    if head is None:
        return None

    prev_last = None
    last = None
    current = head

    # Traverse the linked list to find the last occurrence of the key
    while current.next is not None:
        if current.next.data == key:
            prev_last = current
            last = current.next
        current = current.next

    # Handle the special case when the last occurrence is the head
    if last == head:
        head = head.next
    else:
        prev_last.next = last.next

    return head

# Example usage:
# Linked list: 1 -> 2 -> 3 -> 2 -> 4 -> 2 -> None
head = Node(1)
head.next = Node(2)
head.next.next = Node(3)
head.next.next.next = Node(2)
head.next.next.next.next = Node(4)
head.next.next.next.next.next = Node(2)

key = 2
head = delete_last_occurrence(head, key)

# Output: 1 -> 2 -> 3 -> 2 -> 4 -> None
while head is not None:
    print(head.data, end=" -> ")
    head = head.next
print("None")
```

```python
#6
class Node:
    def __init__(self, data=None):
        self.data = data
        self.next = None

def merge_sorted_lists(a, b):
    dummy = Node()
    tail = dummy

    while a is not None and b is not None:
        if a.data <= b.data:
            tail.next = a
            a = a.next
        else:
            tail.next = b
            b = b.next
        tail = tail.next

    # Attach the remaining nodes from the non-empty list
    if a is not None:
        tail.next = a
    if b is not None:
        tail.next = b

    return dummy.next

# Example usage:
# Linked list a: 5 -> 10 -> 15 -> None
a = Node(5)
a.next = Node(10)
a.next.next = Node(15)

# Linked list b: 2 -> 3 -> 20 -> None
b = Node(2)
b.next = Node(3)
b.next.next = Node(20)

merged_head = merge_sorted_lists(a, b)

# Output: 2 -> 3 -> 5 -> 10 -> 15 -> 20 -> None
while merged_head is not None:
    print(merged_head.data, end=" -> ")
    merged_head = merged_head.next
print("None")
```

```python
#7
class Node:
    def __init__(self, data=None):
        self.data = data
        self.next = None
        self.prev = None

def reverse_doubly_linked_list(head):
    current = head
    previous = None

    while current is not None:
        next_node = current.next
        current.next = previous
        current.prev = next_node
        previous = current
        current = next_node

    head = previous
    head.prev = None

    return head

# Example usage:
# Doubly linked list: 1 <-> 2 <-> 3 <-> 4 <-> None
head = Node(1)
head.next = Node(2)
head.next.prev = head
head.next.next = Node(3)
head.next.next.prev = head.next
head.next.next.next = Node(4)
head.next.next.next.prev = head.next.next

reversed_head = reverse_doubly_linked_list(head)

# Output: 4 <-> 3 <-> 2 <-> 1 <-> None
while reversed_head is not None:
    print(reversed_head.data, end=" <-> ")
    reversed_head = reversed_head.next
print("None")
```

```python
#8
class Node:
    def __init__(self, data=None):
        self.data = data
        self.next = None
        self.prev = None

def delete_node_from_position(head, position):
    if head is None:
        return None

    if position == 0:
        new_head = head.next
        if new_head is not None:
            new_head.prev = None
        return new_head

    current = head
    count = 0

    while current is not None and count < position:
        current = current.next
        count += 1

    if current is None:
        return head

    previous_node = current.prev
    next_node = current.next

    if previous_node is not None:
        previous_node.next = next_node

    if next_node is not None:
        next_node.prev = previous_node

    current.prev = None
    current.next = None

    return head

# Example usage:
# Doubly linked list: 1 <-> 2 <-> 3 <-> 4 <-> None
head = Node(1)
head.next = Node(2)
head.next.prev = head
head.next.next = Node(3)
head.next.next.prev = head.next
head.next.next.next = Node(4)
head.next.next.next.prev = head.next.next

position = 2
new_head = delete_node_from_position(head, position)

# Output: 1 <-> 2 <-> 4 <-> None
while new_head is not None:
    print(new_head.data, end=" <-> ")
    new_head = new_head.next
print("None")
```
